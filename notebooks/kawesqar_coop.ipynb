```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Model functions

## Logistic whaling success

```python
mu_logistic_whaling_success = 0.2
scale=0.1
x=np.arange(0,1,0.01)
for mu in [0.2,0.5,0.8]:
    y=1/(1+np.exp(-(x-mu) / scale))
    plt.plot(x,y,label='mu='+str(mu))
plt.xlabel('Whaling population rate')
plt.ylabel('Probability of success')
plt.legend()
plt.show()
```

## Truncated normal

### Resources

```python
scale=0.1
mu_low=0.2
mu_high=0.8
y_low=np.random.normal(mu_low,scale,1000)
y_high=np.random.normal(mu_high,scale,1000)
plt.hist(y_low,bins=30,density=True,color='red',alpha=0.5,label='Low')
plt.hist(y_high,bins=30,density=True,color='blue',alpha=0.5,label='High')
plt.title('Probability distribution of obtaining resources')
plt.legend()
plt.show()
```

### Social Capital

```python
scale=0.1
mu_low=0.2
mu_medium=0.5
mu_high=0.8
y_low=np.random.normal(mu_low,scale,1000)
y_medium=np.random.normal(mu_medium,scale,1000)
y_high=np.random.normal(mu_high,scale,1000)
plt.hist(y_low,bins=30,density=True,color='red',alpha=0.5,label='Low')
plt.hist(y_medium,bins=30,density=True,color='green',alpha=0.5,label='Medium')
plt.hist(y_high,bins=30,density=True,color='blue',alpha=0.5,label='High')
plt.title('Probability distribution of obtaining social capital')
plt.legend()
plt.show()
```

## Fitness function

```python
theta=0.1
x = np.linspace(0, 1, 100) # resources
y = np.linspace(0, 1, 100) # social capital
X, Y = np.meshgrid(x, y)
fig,axs=plt.subplots(1,3,figsize=(12,4))
for theta,i in zip([0.1,0.5,0.9],range(3)):
    Z = theta*X + (1-theta)*Y # fitness
    axs[i].contourf(X, Y, Z, 100, cmap='RdYlGn')
    axs[i].set_xlabel('Resources')
    axs[i].set_ylabel('Social capital')
    axs[i].set_title('Fitness landscape (theta=%0.1f)' % theta)
plt.show()
```
