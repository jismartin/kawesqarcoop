```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Logistic whaling success

```python
mu_logistic_whaling_success = 0.2
scale=0.1
x=np.arange(0,1,0.01)
y=1/(1+np.exp(-(x-mu_logistic_whaling_success) / scale))
plt.plot(x,y)
plt.plot(mu_logistic_whaling_success,0.5,'ro')
plt.xlabel('Whaling population rate')
plt.ylabel('Probability of success')
plt.show()
```
